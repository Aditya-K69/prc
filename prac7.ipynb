```
import nltk
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

```
document = """
Natural Language Processing is a fascinating field of Artificial Intelligence.
It enables computers to understand human language.
""" 

print("Original Document:")
print(document)
```

```
tokens = word_tokenize(document)
print("Tokens:")
print(tokens)

pos_tags = pos_tag(tokens)
print("\nPOS Tags:")
print(pos_tags)
```

```
stop_words = set(stopwords.words('english'))

filtered_tokens = [
    word.lower()
    for word in tokens
    if word.lower() not in stop_words and word.isalpha()
]

print("Filtered Tokens:")
print(filtered_tokens)
```

```
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_tokens]

print("Stemmed Words:")
print(stemmed_words)

lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]

print("\nLemmatized Words:")
print(lemmatized_words)
```

```
documents = [
    document,
    "Natural language processing enables computers",
    "Computers understand human language"
]
```

```
count_vectorizer = CountVectorizer()
tf_matrix = count_vectorizer.fit_transform(documents)

tf_df = pd.DataFrame(
    tf_matrix.toarray(),
    columns=count_vectorizer.get_feature_names_out()
)

print("Term Frequency (TF):")
print(tf_df)
```

```
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=tfidf_vectorizer.get_feature_names_out()
)

print("TF-IDF Values:")
print(tfidf_df)
```

```
idf_df = pd.DataFrame({
    'Word': tfidf_vectorizer.get_feature_names_out(),
    'IDF Value': tfidf_vectorizer.idf_
})

print("IDF Values:")
print(idf_df)
```

```
print("Vocabulary:")
print(tfidf_vectorizer.get_feature_names_out())
```

```
plt.figure(figsize=(10,5))

sns.heatmap(
    tfidf_df,
    annot=True
)

plt.title("TF-IDF Heatmap")
plt.xlabel("Words")
plt.ylabel("Documents")
plt.show()
```

```
word_scores = tfidf_df.sum(axis=0)

plt.figure(figsize=(10,5))
word_scores.plot(kind='bar')

plt.title("TF-IDF Scores of Words")
plt.xlabel("Words")
plt.ylabel("Score")
plt.show()
```
