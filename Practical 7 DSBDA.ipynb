```python
import nltk
import pandas as pd
import numpy as np

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
sample = """
To be or not to be, that is the question. whether it is nobler for the mind to
suffer the slings and arrows of outrageous fortunes, or to take arms against a sea
of troubles and by opposing, end them. To die, to sleep, perchance have a dream!
Aye, there's the rub!
"""
```

```python
#tokenization
tokens = word_tokenize(sample)
tokens
```

```python
#Stop Words
stop_words = set(stopwords.words('english'))
filtered = [w for w in tokens if w.lower() not in stop_words and w.isalpha()]
filtered
```

```python
#Stemming
stemmer = PorterStemmer()
stemmed = [stemmer.stem(word) for word in filtered]
stemmed
```

```python
#Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(word) for word in filtered]
lemmatized
```

```python
#POS Tagging
pos = pos_tag(filtered)
pos
```

```python
#Term frequency
from sklearn.feature_extraction.text import CountVectorizer

sample2 = ["To be or not to be, that is the question.", 
           "whether it is nobler for the mind to suffer the slings and arrows of outrageous fortunes", 
           "or to take arms against a sea of troubles and by opposing, end them.", 
           "To die, to sleep, perchance have a dream!", "Aye, there's the rub!"]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(sample2)

tf = pd.DataFrame(X.toarray(), columns = vectorizer.get_feature_names_out())

tf
```

```python
#Inverse Document Frequency = importance of words
#common words have low val, rare ones have high
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
X_tfidf = tfidf.fit_transform(sample2)

idf = pd.DataFrame(
    X_tfidf.toarray(),
    columns = tfidf.get_feature_names_out()
)

idf
```
