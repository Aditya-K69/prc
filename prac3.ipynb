```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```
data = {

    'SepalLength':[5.1,4.9,4.7,5.0,5.4,6.4,6.9,5.5,6.5,7.2,6.7,np.nan],
    'SepalWidth':[3.5,3.0,3.2,3.6,3.9,3.2,3.1,2.3,3.0,3.6,20.0,3.0],
    'PetalLength':[1.4,1.4,1.3,1.4,1.7,4.5,4.9,4.0,5.8,6.1,np.nan,5.1],
    'PetalWidth':[0.2,0.2,0.2,0.2,0.4,1.5,1.5,1.3,2.2,2.5,2.1,np.nan],
    'Species':[
        'setosa','setosa','setosa','setosa','Setosa',
        'versicolor','versicolor','versicolor',
        'virginica','virginica','Virginica','virginica'
    ]
}

df = pd.DataFrame(data)

df.to_csv("iris.csv", index=False)
print("CSV Created Successfully")
```

```
print(df.head())
print(df.info())
print(df.shape)
print(df.columns)
```

```
print(df.isnull().sum())
print(df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

```
df['SepalLength'].fillna(df['SepalLength'].median(), inplace=True)
df['PetalLength'].fillna(df['PetalLength'].median(), inplace=True)
df['PetalWidth'].fillna(df['PetalWidth'].median(), inplace=True)

df['Species'] = df['Species'].str.lower()

print(df.isnull().sum())
print(df.dtypes)
print(df.describe())
```

```
plt.figure(figsize=(10,6))
sns.boxplot(data=df)
plt.show()
```

```
grouped = df.groupby('Species')

for species_name, data in grouped:
    print("\nSPECIES:", species_name)
    print(data.mean(numeric_only=True))
    print(data.median(numeric_only=True))
```
