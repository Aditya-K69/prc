```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
data = {
    'Student_ID': [101,102,103,104,105,106,107,108,109,110],

    'Name': [
        'Amit','Priya','Rahul','Sneha','Karan',
        'Pooja','Riya','Arjun','Neha','Vikas'
    ],

    'Gender': [
        'Male','Female','Male','Female','Male',
        'Female','Female','Male','Female','Male'
    ],

    'Maths': [
        78,85,np.nan,92,35,
        88,76,95,500,67
    ],

    'Science': [
        72,90,81,95,40,
        np.nan,79,98,85,65
    ],

    'English': [
        70,82,75,91,38,
        87,74,96,89,-20
    ],

    'Attendance': [
        85,90,88,95,60,
        92,np.nan,97,93,80
    ]
}

df = pd.DataFrame(data)
```

```
print("Original dataset")
print(df)
```

```
print("\nFirst 5 Records:\n")
print(df.head())

print("\nDataset Information:\n")
print(df.info())

print("\nStatistical Summary:\n")
print(df.describe())

print("\nDatatypes:\n")
print(df.dtypes)
```

```
print("\nDuplicate Rows:")
print(df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

```
print("\nMissing Values:\n")
print(df.isnull().sum())
```

```
df['Maths'] = df['Maths'].fillna(df['Maths'].mean())
df['Science'] = df['Science'].fillna(df['Science'].mean())
df['Attendance'] = df['Attendance'].fillna(df['Attendance'].mean())

df.isnull().sum()
```

```
df.loc[df['English'] < 0, 'English'] = df['English'].mean()
df.loc[df['Maths'] > 100, 'Maths'] = df['Maths'].median()
print("\nDataset after handling inconsistencies:\n")
print(df)
```

```
df['Gender'] = df['Gender'].map({
    'Male': 0,
    'Female': 1
})

print("\nDataset after Encoding:\n")
print(df)
```

```
numeric_cols = ['Maths', 'English', 'Science', 'Attendance']

for col in numeric_cols:
    print(f"Checking outliers in {col}")

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3-Q1
    lower_limit = Q1 - 1.15 * IQR
    upper_limit = Q3 + 1.15 * IQR

    outliers = df[
    (df[col] < lower_limit) | (df[col] > upper_limit)
    ]

    print(outliers[col])

    median = df[col].median()

    df.loc[df[col] < lower_limit, col] = median
    df.loc[df[col] > upper_limit, col] = median
```

```
df['Attendance_Log'] = np.log(df['Attendance'])

print("\nDataset after Transformation:\n")
print(df)
```

```
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df[['Maths', 'English', 'Science', 'Attendance']]
)

plt.title("Subplot")
plt.show()
```

```
df[['Maths', 'English', 'Science', 'Attendance']].hist(figsize=(10,5))
plt.show()
```

```
plt.figure(figsize=(8,5))

sns.heatmap(
    data=df[['Maths', 'English', 'Science', 'Attendance']].corr(),
    annot=True,
    cmap='coolwarm'
)

plt.show()
```
