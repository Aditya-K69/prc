```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Iris.csv')
df.head()
```

```python
df.columns
```

```python
#we will group by categorical value 'species' and calculate summay statistics
grouped = df.groupby('Species')['SepalLengthCm'].agg(['mean', 'median', 'min', 'max', 'std'])
grouped
```

```python
#create list of numeric vals for each category
lists = df.groupby('Species')['SepalLengthCm'].apply(list)
lists
```
