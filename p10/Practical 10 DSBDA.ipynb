```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Iris.csv")
df.head()
```

```python
# Features and their types
df.columns
```

```python
df.info()
```

```python
#first four columns = nummerical (continuous)
#last column = nominal (categorical)
```

```python
df = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
#histogram
df.hist(figsize=(10,8), bins=15, edgecolor='black')

# - in histogram, x axis indicates value of variable,and y axis indicates how often
# it occurs in the dataset
```

```python
plt.figure(figsize=(10, 6))

for i, column in enumerate(df.columns, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(y=df[column])
    plt.title(column)

plt.tight_layout()
plt.show()

print(df.columns)
```

```python
#Comparisons:
# - Sepal: wider pread, more overlap between species and has slight outliers in width
# - Petal: clearly seperated distributions, strong clustering by species which is very useful for classifications.

# #Outliers:
# - Some sepal width values fall outside.
# - Petal features show minimum outliers
```
