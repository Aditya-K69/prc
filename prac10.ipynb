```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
data = {

    'SepalLength':[5.1,4.9,4.7,5.0,5.4,6.4,6.9,5.5,6.5,7.2,6.7,np.nan],
    'SepalWidth':[3.5,3.0,3.2,3.6,3.9,3.2,3.1,2.3,3.0,3.6,20.0,3.0],
    'PetalLength':[1.4,1.4,1.3,1.4,1.7,4.5,4.9,4.0,5.8,6.1,np.nan,5.1],
    'PetalWidth':[0.2,0.2,0.2,0.2,0.4,1.5,1.5,1.3,2.2,2.5,2.1,np.nan],
    'Species':[
        'setosa','setosa','setosa','setosa','Setosa',
        'versicolor','Versicolor','versicolor',
        'virginica','virginica','Virginica','virginica'
    ]
}

df = pd.DataFrame(data)
df.to_csv("iris_custom.csv", index=False)

print(df)
```

```
print(df.isnull().sum())

df['Species'] = df['Species'].str.lower()

numeric_cols = ['SepalLength','SepalWidth','PetalLength','PetalWidth']

for col in numeric_cols:
    df[col].fillna(df[col].mean(), inplace=True)

print(df.isnull().sum())
```

```
for col in df.columns:
    if df[col].dtype == 'object':
        print(col, "--> Categorical")
    else:
        print(col, "--> Numeric")

print(df.info())
```

```
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(f"Outliers in {col}:")
    print(outliers[[col]])
```

```
for col in numeric_cols:
    plt.figure()
    plt.hist(df[col])
    plt.title(col)
    plt.show()
```

```
sns.pairplot(df, hue='Species')
plt.show()
```

```
for col in numeric_cols:
    sns.boxplot(y=df[col])
    plt.title(col)
    plt.show()
```

```
for col in numeric_cols:
    sns.boxplot(x='Species', y=col, data=df)
    plt.title(f"{col} by Species")
    plt.show()
```
