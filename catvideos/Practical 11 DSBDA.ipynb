```python
from collections import defaultdict
```

```python
def mapper(text):
    pairs = []
    for word in text.split():
        word = ''.join(c.lower() for c in word if c.isalpha())
        if word:
            pairs.append((word, 1))
    print(pairs)
    return pairs
```

```python
def shuffle(mapped):
    grouped = defaultdict(list)
    for word, count in mapped:
        grouped[word].append(count)
    print(grouped)
    return grouped
```

```python
def reducer(grouped):
    return {
        word: sum(values)
        for word, values in sorted(grouped.items())
    }
    
```

```python
text = "hello there general hello hehe general world"
```

```python
mapped = mapper(text)
grouped = shuffle(mapped)
final = reducer(grouped)
final
```
